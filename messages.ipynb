```python
from langchain_core.messages import BaseMessage

msg = BaseMessage(
    type="Human",
    content="정읍은 어떤 도시인가요?"
)

print(msg)
```

```python
from langchain_core.messages import HumanMessage

msg2 = HumanMessage(
    content="정읍은 어떤 도시인가요?"
)

print(msg2)
print(msg2.type)
```

```python
from langchain_core.messages import SystemMessage

msg3 = SystemMessage(
    content="너는 10년차 시니어 개발자 역할을 수행할거야. 한국어로만 대답해."
)

print(msg3)
print(msg3.type)
```
